```python
def plot_equation():
  import numpy as np
  import matplotlib.pyplot as plt
  x=np.linspace(-100,100,200)
  n=int(input("Enter the value"))
  y=[i**n for i in x]
  plt.plot(x,y)
  plt.title("LineChart")
  plt.xlabel("X-axis")
  plt.ylabel("Y-axis")
  plt.show()
```
