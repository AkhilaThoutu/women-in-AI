```python
message="Encrypt this message using python!"
encrypted_message=""
for i in message:
  if i.isalpha():
    encrypted_message+=chr(219-ord(i))
  elif i==" ":
    encrypted_message+="**"
  else:
    encrypted_message+=i
print(encrypted_message)
```

```python
message="Encrypt this using python!"
encrypted_message=""
for i in message.lower():
  if i.isalpha():
    encrypted_message+=chr(219-ord(i))
  elif i==" ":
    encrypted_message+="*"
  else:
    encrypted_message+=i
print(encrypted_message)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv('/content/Position Salary .csv')
```

```python
df.head(5)
```

```python
x=df.iloc[:,1:2].values
print(x)
```

```python
y=df.iloc[:,2].values
print(y)
```

```python
plt.plot(x,y)
plt.show()
```

```python
from sklearn.tree import DecisionTreeRegressor
dt=DecisionTreeRegressor()
dt.fit(x,y)
print(dt)
```

```python
y_pred=dt.predict(x)
print(y_pred)
```

```python
r2_score(y,y_pred)
```

```python
from sklearn.linear_model import LinearRegression
l=LinearRegression()
l=l.fit(x,y)
l
```

```python
y_pred=l.predict(x)
y_pred
```

```python
r2_score(y,y_pred)
```
