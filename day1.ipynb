```python
intlist=[]
strlist=[]
floatlist=[]
for i in list2:
  if type(i)==str:
    strlist.append(i)
  elif type(i)==int:
    intlist.append(i)
  else:
    floatlist.append(i)
print("intlist=",intlist,"strlist=",strlist,"floatlist=",floatlist)
```

```python
#list comprehension
intlist=[i for i in list2 if type(i)==int]
strlist=[i for i in list2 if type(i)==str]
floatlist=[i for i in list2 if type(i)==float]
print(intlist,strlist,floatlist)
```

```python
#create a function of list that will separate elements of the list
def separate(x):
  intlist=[i for i in x if type(i)==int]
  strlist=[i for i in x if type(i)==str]
  floatlist=[i for i in x if type(i)==float]
  return intlist,strlist,floatlist
y=[12,"thyme",56.2,56]
print(separate(y))
sep_list=separate(y)
for i in sep_list:
  print(i)
```

```python
#Reccurssive function
def recursum(n):
  if n<=1:
    return n
  return n + recursum(n-1)
n=10
print(recursum(n))
```

```python
def recur_fib(n):
  if n<=1:
    return (n)
  else:
    return(recur_fib(n-1)+recur_fib(n-2))
n=5
for i in range(n):
  print(recur_fib(i))
```

```python
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n-1)
i = 5
print( i, factorial(i))
```

```python
def gen_sum(n):
  v=0
  while v<n:
    yield v
    v+=1
for v in gen_sum(100):
  print(v)
```
